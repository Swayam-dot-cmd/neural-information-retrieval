# Final Comparison: BM25 vs BGE vs HyDE

This notebook presents a comparative analysis of lexical, dense, and query expansion-based retrieval methods on the BEIR SciFact dataset.

```python
bm25 = 0.6280555555555555
bge = 0.7985555555555556
hyde = 0.8035555555555556

print("BM25:", bm25)
print("BGE:", bge)
print("HyDE:", hyde)
```

```python
import matplotlib.pyplot as plt

methods = ['BM25', 'BGE', 'HyDE']
scores = [bm25, bge, hyde]

plt.figure()
plt.bar(methods, scores)

plt.title("Recall@10 Comparison")
plt.ylabel("Score")

plt.show()
```

## Observations

- Dense retrieval (BGE) significantly outperforms BM25.
- HyDE further improves performance over BGE, but only marginally.
- This indicates that semantic retrieval is more effective than lexical matching.

## Analysis: Why is HyDE Improvement Small?

Although HyDE improves performance, the improvement is limited.

Reasons:

1. **Quality of Generation Model**
   - HyDE relies on generating hypothetical documents.
   - We used a small model (Flan-T5-small), which produces lower-quality text compared to large LLMs used in the original paper.

2. **Strong Baseline (BGE)**
   - BGE already captures semantic similarity effectively.
   - When the baseline is strong, further improvements become difficult.

3. **Dataset Characteristics**
   - SciFact is a scientific dataset requiring precise terminology.
   - Generated text may introduce noise or slight inaccuracies.

4. **Noise in Generated Content**
   - HyDE may add irrelevant or redundant information.
   - This can sometimes reduce retrieval quality instead of improving it.

## Key Insight

Query expansion techniques like HyDE are not universally beneficial.

Their effectiveness depends on:
- Quality of the generation model
- Nature of the dataset
- Strength of the baseline retrieval model

In this case, HyDE provides only marginal improvement due to limitations of the generation model.

## Conclusion

- BM25 performs well for keyword-based retrieval but lacks semantic understanding.
- BGE significantly improves performance by capturing semantic similarity.
- HyDE provides a slight improvement, but its effectiveness is limited by generation quality.

Overall, dense retrieval remains the most reliable approach in this setup.
