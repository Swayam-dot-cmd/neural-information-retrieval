# Comparison of BM25 vs Dense Retrieval (BGE)

This notebook compares classical lexical retrieval (BM25) with dense embedding-based retrieval (BGE) using Recall@10 on the BEIR SciFact dataset.

```python
bm25_score = 0.6280555555555555
bge_score = 0.7985555555555556

print("BM25 Recall@10:", bm25_score)
print("BGE Recall@10:", bge_score)
```

```python
import matplotlib.pyplot as plt

methods = ['BM25', 'BGE']
scores = [bm25_score, bge_score]

plt.figure()
plt.bar(methods, scores)

plt.title("Recall@10 Comparison")
plt.ylabel("Score")

plt.show()
```

## Observations

- Dense retrieval using BGE significantly outperforms BM25.
- BGE achieves higher Recall@10, indicating better retrieval of relevant documents.
- This suggests that semantic similarity is more effective than keyword matching for this dataset.

## Analysis

BM25 relies on exact keyword matching and term frequency, which makes it effective for queries containing specific terms.

However, it struggles when:
- Synonyms are used
- Query wording differs from document wording

BGE, being a dense embedding model, captures semantic meaning and retrieves documents based on similarity in embedding space.

This allows it to:
- Handle paraphrased queries
- Capture contextual meaning
- Improve retrieval performance significantly

## Conclusion

Dense retrieval methods outperform classical lexical methods on the SciFact dataset.

This demonstrates the importance of semantic understanding in modern information retrieval systems.
