```python
bm25 = 0.628
bge = 0.799
hyde = 0.804
hybrid = 0.819

print("BM25:", bm25)
print("BGE:", bge)
print("HyDE:", hyde)
print("Hybrid:", hybrid)
```

```python
import matplotlib.pyplot as plt

methods = ['BM25', 'BGE', 'HyDE', 'Hybrid']
scores = [bm25, bge, hyde, hybrid]

plt.figure()
plt.bar(methods, scores)
plt.title("Final Comparison")
plt.ylabel("Recall@10")
plt.show()
```

## Final Insights

- Dense retrieval significantly outperforms BM25.
- HyDE provides marginal improvements, limited by generation quality.
- Hybrid retrieval achieves the best performance.

### Key Observation:
A small contribution from BM25 improves performance over pure dense retrieval.

This shows that lexical and semantic methods are complementary.
